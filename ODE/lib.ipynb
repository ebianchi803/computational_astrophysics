```python
from scipy.optimize import root
```

```python
def euler_explicit(t,y,df_dt,h,args=None):
    ''' Forward Euler method
        ...
    '''
    y_new = y + h*df_dt(t,y,args)
    
    return y_new
```

```python
def euler_implicit(t,y,df_dt,h,args=None):
    ''' Backwards Euler method
        ...
    '''
    
    y_new = (root(lambda yy : yy - y - h*df_dt(t+h,yy,args), y)).x
    y_new = y + h*df_dt(t+h,y_new,args)

    return y_new
```

```python
def midpoint_explicit(t,y,df_dt,h,args=None):
    ''' Explicit Midpoint method
        ...
    '''
    y_new_2 = y + h/2*df_dt(t,y,args)
    y_new = y + h*df_dt(t+h/2,y_new_2,args)
    
    return y_new
```

```python
def midpoint_implicit(t,y,df_dt,h,args=None):
    ''' Implicit Midpoint method
        ...
    '''
    y_new = (root(lambda yy : yy - y - h*df_dt(t+h,y,args), y)).x
    y_new = y + h*df_dt(t+h/2,(y_new+y)/2,args)
    
    return y_new
```

```python
def predictor_corrector(t,y,df_dt,h,args=None):
    ''' predictor corrector method
        ...
    '''
    y_new_P = y + h*df_dt(t,y,args)
    y_new = y + h/2*(df_dt(t,y,args) + df_dt(t+h,y_new_P,args))

    return y_new
```

```python
def RK4(t,y,df_dt,h,args=None):
    ''' Order 4 Runge Kutta method
        ...
    '''
    k1 = df_dt(t,y,args)
    k2 = df_dt(t+h/2,y+k1*h/2,args)
    k3 = df_dt(t+h/2,y+k2*h/2,args)
    k4 = df_dt(t+h,y+k3*h,args)

    y_new = y + h/6*(k1+2*k2+2*k3+k4)
    return y_new
```

```python
def integrate(df_dt, t_range, IC, h, args=None, method=None):
    
    t = t_range[0]
    y = np.array(IC)
    t_array = [t]
    y_array = [y]
    while t < t_range[1]:
        y_new = method(t, y, df_dt, h, args)
        t += h
        y = np.array(y_new)
        t_array.append(t)
        y_array.append(y)
    t_array, y_array = np.array(t_array),np.array(y_array)
    y_array = y_array.T
    return t_array, y_array
```
